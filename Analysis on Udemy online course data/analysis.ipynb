```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Assuming you have loaded your dataset into 'data' DataFrame and identified numerical features
numerical_features = [
    'avg_rating','discount_price__amount',
    'id', 'num_published_lectures',
    'num_published_practice_tests', 'price_detail__amount',
    'num_reviews', 'rating', 'avg_rating_recent'
]

# Extract X and y from your dataset
X = data[numerical_features]
y = data['num_subscribers']

# Step 1: Handle NaN values in the input features (X)
# Option 1: Impute NaN values in X
imputer_X = SimpleImputer(strategy='mean')
X = imputer_X.fit_transform(X)

# Step 2: Handle NaN values in the target variable (y)
imputer_y = SimpleImputer(strategy='mean')
y = imputer_y.fit_transform(y.values.reshape(-1, 1)).ravel()

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

# Step 4: Perform feature scaling (standardization) on numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 5: Fit PolynomialFeatures and Ridge regression model
degree = 2  # Example polynomial degree
poly = PolynomialFeatures(degree)
X_train_pf = poly.fit_transform(X_train_scaled)
X_test_pf = poly.transform(X_test_scaled)

poly_linear = Ridge(fit_intercept=True, random_state=42).fit(
    X_train_pf, y_train)

# Step 6: Evaluate the model
scr = poly_linear.score(X_test_pf, y_test)
mse = mean_squared_error(y_test, poly_linear.predict(X_test_pf))
mae = mean_absolute_error(y_test, poly_linear.predict(X_test_pf))

# Collect performance statistics
poly_stats = {}
poly_stats['model_name'] = 'Polynomial Regression'
poly_stats['min_coef'] = float('%.3f' % poly_linear.coef_.min())
poly_stats['max_coef'] = float('%.3f' % poly_linear.coef_.max())
poly_stats['mean_abs_coef'] = float('%.3f' % np.abs(poly_linear.coef_).mean())
poly_stats['num_zeros'] = np.sum(poly_linear.coef_ == 0)
poly_stats['test_score'] = float('%.3f' % scr)
poly_stats['MSE'] = float('%.3f' % mse)
poly_stats['MAE'] = float('%.3f' % mae)

# Print performance metrics
print(f'R-squared score: {scr:.3f}')
print(f'Mean Absolute Error: {mae:.3f}')
print(f'Mean Squared Error: {mse:.3f}')

# Print additional statistics
print(f"Minimum Coefficient: {poly_stats['min_coef']}")
print(f"Maximum Coefficient: {poly_stats['max_coef']}")
print(f"Mean Absolute Coefficient: {poly_stats['mean_abs_coef']}")
print(f"Number of Zero Coefficients: {poly_stats['num_zeros']}")

# Append the statistics to a list or dictionary
performance.append(poly_stats)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error

# Load the dataset
data = pd.read_csv(
    'C:/Users/shrey/OneDrive/Desktop/data analyst task/udemy_output_All_Finance__Accounting_p1_p626.csv')

# Display basic information about the dataset
print(data.info())
```

```python
# Drop rows with missing values and empty values
data.dropna(inplace=True)

# Separate features (X) and target (y)
# Drop non-numeric and non-predictive columns
X = data.drop(['url', 'title'], axis=1)
y = data['num_subscribers']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
# Define numerical and categorical features
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object', 'category']).columns

# Preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Fit and transform preprocessing pipeline on training data
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

# 1. Random Forest Regression

```python
# Initialize Random Forest Regressor
rf_reg = RandomForestRegressor(max_depth=4, random_state=42)

# Train the model
rf_reg.fit(X_train_preprocessed, y_train)

# Evaluate the model
rf_scr = rf_reg.score(X_test_preprocessed, y_test)
rf_mse = mean_squared_error(y_test, rf_reg.predict(X_test_preprocessed))
rf_mae = mean_absolute_error(y_test, rf_reg.predict(X_test_preprocessed))

# Print evaluation metrics
print("Random Forest Regressor:")
print(f"R-squared score: {rf_scr}")
print(f"Mean Absolute Error: {rf_mae}")
print(f"Mean Squared Error: {rf_mse}")
```

# 2. XGBoost Regressor

```python
# Initialize XGBoost Regressor
xgb_reg = XGBRegressor(objective='reg:squarederror', random_state=42)

# Train the model
xgb_reg.fit(X_train_preprocessed, y_train)

# Evaluate the model
xgb_scr = xgb_reg.score(X_test_preprocessed, y_test)
xgb_mse = mean_squared_error(y_test, xgb_reg.predict(X_test_preprocessed))
xgb_mae = mean_absolute_error(y_test, xgb_reg.predict(X_test_preprocessed))

# Print evaluation metrics
print("/nXGBoost Regressor:")
print(f"R-squared score: {xgb_scr}")
print(f"Mean Absolute Error: {xgb_mae}")
print(f"Mean Squared Error: {xgb_mse}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the dataset
file_path = 'C:/Users/shrey/OneDrive/Desktop/data analyst task/udemy_output_All_Finance__Accounting_p1_p626.csv'
df = pd.read_csv(file_path)

# Displaying the first few rows and columns of the dataset
print(df.head())

# Step 2: Data Preprocessing
# Handle missing values
imputer = SimpleImputer(strategy='mean')
df[numerical_columns] = imputer.fit_transform(df[numerical_columns])

# Encode categorical variables (if any)
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(
    lambda col: le.fit_transform(col))

# Step 3: Compute the correlation matrix
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numerical_columns].corr()

# Step 4: Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu',
            fmt='.2f', square=True, cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
